# Data Analysis of UK Divorce Rates 

```python
# Import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Import Dataset 

df = pd.read_csv("UKDivorceAndMarriage.csv")
df.head()
```

```python
# Cleaning process 1: 
# Removing the commas and converting the values to floats 

df['Total Number of Divorces'] = df['Total Number of Divorces'].str.replace(',', '').astype(float)
```

```python
#Result
df.head(10)
```

### Total Number of Divorce Rates

```python
# By using the iloc method, I'm able to select the relevant timeframe 

x = df.loc[0:6]['Year']
y = df.loc[0:6]['Total Number of Divorces']
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Total Number of Divorce Rates')
plt.title('Divorce rates within a 5 year period')

# This particular block of code was used to further improve the image quality of the graph 
%config InlineBackend.figure_format = 'svg'
```

### Analysis 1:

```python
# Based on results, it's clear that the overall divorce rates in the UK decreeased drastically (up to 15%) over the 10 year period, 
# which also shows a optmistic outlook on future marriages. 
# However, the limitations of the dataset being somewhat outdated (i.e. by 4 years), when further analysing this model.
```

```python
# Cleaning process 2:
#Removing colons and commas for the Husband and Wife columns

df['Husband Submitted Divorce'] = df['Husband Submitted Divorce'].replace(':',0)
df['Husband Submitted Divorce'] = df['Husband Submitted Divorce'].str.replace(',', '').astype(float)

df['Wife Submitted Divorce'] = df['Wife Submitted Divorce'].replace(':',0)
df['Wife Submitted Divorce'] = df['Wife Submitted Divorce'].str.replace(',', '').astype(float)

#Drop all NaN Null Values, to prevent potential errors that can affect analysis 

df.dropna(inplace = True)
```

### Divorce Submission Rates

```python
#Comparison of Divroce submissions through seperate bar charts

df.iloc[0:10].plot(kind = 'bar', x='Year', y='Husband Submitted Divorce', title = 'Male Divorce Submissions within a 10 year period')
plt.gca().invert_xaxis()
df.iloc[0:10].plot(kind = 'bar', x='Year', y='Wife Submitted Divorce', title = 'Female Divorce Submissions within a 10 year period')
plt.gca().invert_xaxis()
%config InlineBackend.figure_format = 'svg'
```

```python
#ALternative method with both charts being stacked toghter to show a much more clear model for analysis

%matplotlib tk
df1=df[1:11]
df1.plot('Year',['Husband Submitted Divorce','Wife Submitted Divorce'],kind = 'bar')
plt.ylabel('Divorce Submissions')
```

### Analysis 2:

```python
# Based on the results, it's clear that female/wife divorce submissions are 
# significantly higher than the divroce submissions by the men/husband
# Hence, majority of the divorves over the past 10 years have been initated by the wife
```

### Medium Age of Marriage

```python
#Graph from 1991-2010

%matplotlib tk
df1=df[0:51]
df1.plot('Year',['Median Age of Single Males Marriage','Median Age of Single Females Marriage'],kind = 'bar')
plt.ylabel('Medium Age of Marriage')
plt.legend(bbox_to_anchor=(1, 1.08), loc=7)

```

### Analysis 3:

```python
# Based from the graphs it's clear that the medium age for marriage has increased
# The increase is more significant for women and this could be due a number of reasons
# For example, women nowadays are much more career driven and this could explain why- 
# women delay marriage to a later age. 
# The reasons can also apply to men. 
```

### Re-marriage Rates 

```python
x = df.loc[1:6]['Year']
y = df.loc[1:6]['Number Men Re-marrying per 1,000 single population aged 16 and over']
plt.bar(x,y, label='Men')
x = df.loc[1:6]['Year']
y = df.loc[1:6]['Number Women Re-marrying per 1,000 single population aged 16 and over']
plt.xlabel('Year'); plt.ylabel('Rate of Re-marriage')
plt.title('Rate of Re-marriage For Both Men and Women over a 5 Year Period')
plt.bar(x,y, label='Women')
plt.legend(bbox_to_anchor=(1, 0.7), loc=7)
%config InlineBackend.figure_format = 'svg'
```
